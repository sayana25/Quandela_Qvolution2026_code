What the data actually is:

Each row = one trading day, each column = implied volatility for a specific (Tenor, Maturity) pair
224 vol values per day,

forming a volatility surface — a 14×16 grid

Level 1 (494 days) = training, Level 2 (489 days) = test

The actual task: predict tomorrow's full vol surface given today's. This is next-day surface forecasting — a regression problem with 224 outputs.

```python
# =============================================================================
# Quantum Reservoir Computing — MerLin v4 — Swaption Vol Surface Forecasting
# =============================================================================
#
# Dataset : Quandela/Challenge_Swaptions (Qvolution hackathon)
# Task    : Next-day implied volatility surface prediction
#
# Pipeline
# --------
#   1. Load level-1 (train) and level-2 (test) vol surface data
#   2. PCA compress: 224-dim surface -> N_MODES latent scores
#      (vol surfaces are smooth; first 4 PCs explain ~98% of variance)
#   3. Build lag pairs: X[t] = today's scores, y[t+1] = tomorrow's scores
#   4. Frozen QRC reservoir: N_MODES-dim -> Fock probability features
#   5. Linear readout: Fock features -> N_MODES-dim predicted scores
#   6. PCA reconstruct: predicted scores -> full 224-dim surface
#   7. Evaluate with RMSE and R² on reconstructed surface
#   8. Visualize predicted vs actual surface slices
#
# Why this works
# --------------
#   - Vol surfaces live on a low-dimensional manifold -> PCA gives clean inputs
#   - QRC maps smooth continuous inputs to polynomial Fock features
#   - Linear readout on Fock features approximates the next-day dynamics
#   - This is much stronger than the Iris classification experiment because
#     the target function (vol surface evolution) is a smooth nonlinear map
#     of a handful of latent variables — exactly what QRC is designed for
#
# =============================================================================

import warnings
warnings.filterwarnings("ignore")

import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

from datasets import load_dataset
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import r2_score

from merlin import MeasurementStrategy, QuantumLayer
from merlin.builder import CircuitBuilder

# ── Seeds ─────────────────────────────────────────────────────────────────────
SEED = 42
torch.manual_seed(SEED)
np.random.seed(SEED)

# =============================================================================
# 1) Load data
# =============================================================================
print("Loading datasets...")

ds_train = load_dataset(
    "Quandela/Challenge_Swaptions",
    data_files="level-1_Future_prediction/train.csv",
    split="train",
)
ds_test = load_dataset(
    "Quandela/Challenge_Swaptions",
    data_files="level-2_Missing_data_prediction/train_level2.csv",
    split="train",
)

import pandas as pd

df_train = ds_train.to_pandas()
df_test  = ds_test.to_pandas()

# Drop the Date column — we only use the vol surface values
VOL_COLS = [c for c in df_train.columns if c != "Date"]

X_train_raw = df_train[VOL_COLS].values.astype("float32")  # (494, 224)
X_test_raw  = df_test[VOL_COLS].values.astype("float32")   # (489, 224)

# Handle any NaNs in test set by column-mean imputation
col_means = np.nanmean(X_train_raw, axis=0)
for j in range(X_test_raw.shape[1]):
    mask = np.isnan(X_test_raw[:, j])
    if mask.any():
        X_test_raw[mask, j] = col_means[j]

print(f"  Train surface shape : {X_train_raw.shape}")
print(f"  Test  surface shape : {X_test_raw.shape}")

# =============================================================================
# 2) PCA compression: 224-dim -> N_MODES latent scores
#    Fit on train only, apply to both sets.
# =============================================================================
N_MODES     = 6    # number of PCA components = quantum modes
DEPTH       = 5     # reservoir depth
N_PHOTONS   = 3     # Fock space dimension = C(N_MODES+N_PHOTONS, N_PHOTONS)

pca = PCA(n_components=N_MODES, random_state=SEED)
pca.fit(X_train_raw)

var_explained = pca.explained_variance_ratio_.sum()
print(f"\n  PCA variance explained by {N_MODES} components: {var_explained:.2%}")

scores_train = pca.transform(X_train_raw)  # (494, 4)
scores_test  = pca.transform(X_test_raw)   # (489, 4)

# Standardize the latent scores (fit scaler on train)
scaler = StandardScaler()
scores_train = scaler.fit_transform(scores_train).astype("float32")
scores_test  = scaler.transform(scores_test).astype("float32")

# =============================================================================
# 3) Build lag pairs for next-day forecasting
#    X[t] = today's PCA scores  ->  y[t] = tomorrow's PCA scores
# =============================================================================
def make_lag_pairs(scores):
    """Return (X, y) where X[i] = scores[i], y[i] = scores[i+1]."""
    X = torch.tensor(scores[:-1], dtype=torch.float32)
    y = torch.tensor(scores[1:],  dtype=torch.float32)
    return X, y

X_tr, y_tr = make_lag_pairs(scores_train)  # (493, 4), (493, 4)
X_te, y_te = make_lag_pairs(scores_test)   # (488, 4), (488, 4)

print(f"\n  Training pairs : {X_tr.shape[0]}")
print(f"  Test pairs     : {X_te.shape[0]}")

# =============================================================================
# 4) Quantum Reservoir
#
#   Single angle-encoding block at input (name="enc", N_MODES angles).
#   Reservoir follows Clements order: superpositions -> rotations -> entangling.
#   All reservoir weights are trainable=True (registered as nn.Parameters),
#   then frozen with requires_grad_(False).
#   input_size = N_MODES = 4, one angle per optical mode.
# =============================================================================
def build_reservoir(
    n_photons:      int                 = N_PHOTONS,
    encoding_scale: float               = np.pi / 3,
    measurement:    MeasurementStrategy = None,
    seed:           int                 = SEED,
):
    if measurement is None:
        measurement = MeasurementStrategy.probs()

    torch.manual_seed(seed)
    np.random.seed(seed)

    builder = CircuitBuilder(n_modes=N_MODES)

    builder.add_angle_encoding(
        modes=list(range(N_MODES)),
        name="enc",
        scale=encoding_scale,
    )

    for d in range(DEPTH):
        builder.add_superpositions(depth=1, trainable=True, name=f"bs_{d}")
        builder.add_rotations(trainable=True,               name=f"phi_{d}")
        builder.add_entangling_layer(trainable=True,        name=f"ent_{d}")

    layer = QuantumLayer(
        input_size=N_MODES,
        builder=builder,
        n_photons=n_photons,
        measurement_strategy=measurement,
    )

    for p in layer.parameters():
        p.requires_grad_(False)

    return layer


def extract_features(reservoir, X):
    """Map (batch, N_MODES) -> (batch, fock_dim) Fock probability features."""
    with torch.no_grad():
        q = reservoir(X)
    if torch.is_complex(q):
        return torch.cat([q.real, q.imag], dim=1).float()
    return q.float()


# =============================================================================
# 5) Linear readout for regression
#    Input  : Fock probability features
#    Output : N_MODES predicted PCA scores for the next day
# =============================================================================
def make_readout(feat_dim, out_dim=N_MODES):
    return nn.Linear(feat_dim, out_dim)

def train_readout(model, X_feat, y, lr=0.01, epochs=3000, patience=100):
    criterion = nn.MSELoss()
    opt       = torch.optim.Adam(model.parameters(), lr=lr)
    sched     = torch.optim.lr_scheduler.CosineAnnealingLR(opt, T_max=epochs)

    best_loss, wait, best_state = float("inf"), 0, None
    history = []

    for epoch in range(epochs):
        model.train()
        opt.zero_grad()
        pred = model(X_feat)
        loss = criterion(pred, y)
        loss.backward()
        opt.step()
        sched.step()

        lv = loss.item()
        history.append(lv)

        if lv < best_loss - 1e-7:
            best_loss, wait = lv, 0
            best_state = {k: v.clone() for k, v in model.state_dict().items()}
        else:
            wait += 1
            if wait >= patience:
                break

    if best_state:
        model.load_state_dict(best_state)
    return history

def predict_surface(model, reservoir, X_scores):
    """
    Given standardized PCA scores X_scores (n, N_MODES), predict next-day
    scores and reconstruct the full vol surface.
    """
    feats       = extract_features(reservoir, X_scores)
    with torch.no_grad():
        pred_scores = model(feats).numpy()

    # Inverse-transform: un-standardize then inverse-PCA
    pred_unscaled = scaler.inverse_transform(pred_scores)
    pred_surface  = pca.inverse_transform(pred_unscaled)
    return pred_surface, pred_scores

def surface_metrics(pred, true):
    rmse = np.sqrt(np.mean((pred - true) ** 2))
    mae  = np.mean(np.abs(pred - true))
    r2   = r2_score(true.flatten(), pred.flatten())
    return {"rmse": rmse, "mae": mae, "r2": r2}


# =============================================================================
# 6) Run experiment: compare 4 reservoir configs
#
#   Config A -- baseline:  n_photons=2, scale=1.0,   AMPLITUDES
#   Config B -- +photons:  n_photons=3, scale=1.0,   AMPLITUDES
#   Config C -- +scale:    n_photons=3, scale=pi/3,  AMPLITUDES
#   Config D -- +probs:    n_photons=3, scale=pi/3,  probs()
# =============================================================================
configs = {
    "A: baseline": dict(n_photons=2, encoding_scale=1.0,
                        measurement=MeasurementStrategy.AMPLITUDES),
    "B: +photons":  dict(n_photons=3, encoding_scale=1.0,
                        measurement=MeasurementStrategy.AMPLITUDES),
    "C: +scale":    dict(n_photons=3, encoding_scale=np.pi / 3,
                        measurement=MeasurementStrategy.AMPLITUDES),
    "D: +probs":    dict(n_photons=3, encoding_scale=np.pi / 3,
                        measurement=MeasurementStrategy.probs()),
}

# Also run a classical baseline: linear regression directly on PCA scores
# (no quantum reservoir — just predict tomorrow's scores from today's)
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=1.0)
ridge.fit(scores_train[:-1], scores_train[1:])
ridge_pred_scores = ridge.predict(scores_test[:-1])
ridge_pred_surface = pca.inverse_transform(scaler.inverse_transform(ridge_pred_scores))
true_surface_test  = pca.inverse_transform(scaler.inverse_transform(scores_test[1:]))
ridge_metrics = surface_metrics(ridge_pred_surface, true_surface_test)
print(f"\n  Classical Ridge baseline:"
      f"  RMSE={ridge_metrics['rmse']:.6f}"
      f"  R2={ridge_metrics['r2']:.4f}")

results = {}

for cfg_name, cfg_kwargs in configs.items():
    print(f"\n{'='*60}")
    print(f"  Config {cfg_name}")
    print(f"{'='*60}")

    reservoir = build_reservoir(**cfg_kwargs)
    print(f"  Photons={cfg_kwargs['n_photons']}  "
          f"scale={cfg_kwargs['encoding_scale']:.3f}  "
          f"output_size={reservoir.output_size}")

    # Extract Fock features
    F_tr = extract_features(reservoir, X_tr)
    F_te = extract_features(reservoir, X_te)
    feat_dim = F_tr.shape[1]
    print(f"  Fock feature dim: {feat_dim}")

    # Train linear readout
    readout = make_readout(feat_dim)
    history = train_readout(readout, F_tr, y_tr)

    # Predict on test set
    readout.eval()
    with torch.no_grad():
        pred_scores_test = readout(F_te).numpy()

    pred_surface = pca.inverse_transform(scaler.inverse_transform(pred_scores_test))
    m = surface_metrics(pred_surface, true_surface_test)

    # Also evaluate on train set
    with torch.no_grad():
        pred_scores_train = readout(F_tr).numpy()
    pred_surface_train = pca.inverse_transform(
        scaler.inverse_transform(pred_scores_train))
    true_surface_train = pca.inverse_transform(
        scaler.inverse_transform(scores_train[1:]))
    m_tr = surface_metrics(pred_surface_train, true_surface_train)

    print(f"  Train  RMSE={m_tr['rmse']:.6f}  R2={m_tr['r2']:.4f}")
    print(f"  Test   RMSE={m['rmse']:.6f}  R2={m['r2']:.4f}")

    vs_ridge = "better" if m["rmse"] < ridge_metrics["rmse"] else "worse"
    print(f"  vs Ridge baseline: {vs_ridge} "
          f"(delta RMSE {m['rmse']-ridge_metrics['rmse']:+.6f})")

    results[cfg_name] = dict(
        metrics_tr=m_tr, metrics_te=m,
        history=history,
        pred_surface=pred_surface,
        feat_dim=feat_dim,
    )


# =============================================================================
# 7) Visualization Dashboard
# =============================================================================
plt.rcParams.update({
    "font.family":      "monospace",
    "axes.facecolor":   "#0d1117",
    "figure.facecolor": "#0a0c12",
    "axes.edgecolor":   "#2a2d3e",
    "axes.labelcolor":  "#94a3b8",
    "xtick.color":      "#64748b",
    "ytick.color":      "#64748b",
    "text.color":       "#e2e8f0",
    "grid.color":       "#1a1d2e",
    "grid.linewidth":   0.6,
})

COLORS   = {"A: baseline": "#64748b", "B: +photons": "#f59e0b",
            "C: +scale":   "#a78bfa", "D: +probs":   "#00e5ff"}
cfg_keys = list(results.keys())
cols     = list(COLORS.values())

fig = plt.figure(figsize=(24, 22))
fig.patch.set_facecolor("#0a0c12")
gs  = gridspec.GridSpec(4, 4, figure=fig, hspace=0.55, wspace=0.40)

fig.text(0.5, 0.978,
         "QRC v4 -- Swaption Vol Surface Forecasting -- MerLin",
         ha="center", va="top", fontsize=16, fontweight="bold", color="#e2e8f0")
fig.text(0.5, 0.962,
         f"PCA {N_MODES} components ({var_explained:.1%} variance)  "
         f"| {N_MODES} modes  {N_PHOTONS} photons  depth {DEPTH}  "
         f"| next-day surface prediction",
         ha="center", va="top", fontsize=9, color="#64748b")

# ── Row 0: RMSE bar chart ─────────────────────────────────────────────────────
ax0 = fig.add_subplot(gs[0, :2])
rmse_tr = [results[k]["metrics_tr"]["rmse"] for k in cfg_keys]
rmse_te = [results[k]["metrics_te"]["rmse"] for k in cfg_keys]
x_pos   = np.arange(len(cfg_keys))
w       = 0.35
b_tr = ax0.bar(x_pos - w/2, rmse_tr, width=w, color=cols, alpha=0.9,  label="Train RMSE")
b_te = ax0.bar(x_pos + w/2, rmse_te, width=w, color=cols, alpha=0.45, label="Test RMSE")
for bar, val in zip(list(b_tr) + list(b_te), rmse_tr + rmse_te):
    ax0.text(bar.get_x() + bar.get_width() / 2, val + 1e-5,
             f"{val:.5f}", ha="center", va="bottom", fontsize=7,
             fontweight="bold", color="#e2e8f0")
ax0.axhline(ridge_metrics["rmse"], color="#ef4444", lw=1.2, ls="--",
            label=f"Ridge baseline {ridge_metrics['rmse']:.5f}")
ax0.set_xticks(x_pos); ax0.set_xticklabels(list(COLORS.keys()), fontsize=9)
ax0.set_ylabel("RMSE (implied vol units)")
ax0.set_title("Config Comparison -- Surface RMSE", fontsize=11, color="#e2e8f0", pad=8)
ax0.legend(framealpha=0.15, facecolor="#1a1d2e", fontsize=8)
ax0.grid(axis="y", alpha=0.3); ax0.set_facecolor("#0d1117")

# ── Row 0 right: R² bar chart ─────────────────────────────────────────────────
ax0b = fig.add_subplot(gs[0, 2:])
r2_te = [results[k]["metrics_te"]["r2"] for k in cfg_keys]
b_r2  = ax0b.bar(x_pos, r2_te, color=cols, width=0.5, alpha=0.85)
for bar, val in zip(b_r2, r2_te):
    ax0b.text(bar.get_x() + bar.get_width() / 2, val + 0.002,
              f"{val:.4f}", ha="center", va="bottom", fontsize=9,
              fontweight="bold", color="#e2e8f0")
ax0b.axhline(ridge_metrics["r2"], color="#ef4444", lw=1.2, ls="--",
             label=f"Ridge R2 {ridge_metrics['r2']:.4f}")
ax0b.axhline(1.0, color="#22c55e", lw=0.8, ls=":", alpha=0.4)
ax0b.set_xticks(x_pos); ax0b.set_xticklabels(list(COLORS.keys()), fontsize=9)
ax0b.set_ylabel("R2 score"); ax0b.set_ylim(max(0, min(r2_te) - 0.05), 1.05)
ax0b.set_title("Test R2 -- Config Comparison", fontsize=11, color="#e2e8f0", pad=8)
ax0b.legend(framealpha=0.15, facecolor="#1a1d2e", fontsize=8)
ax0b.grid(axis="y", alpha=0.3); ax0b.set_facecolor("#0d1117")

# ── Row 1: training loss curves ───────────────────────────────────────────────
ax1 = fig.add_subplot(gs[1, :2])
for i, k in enumerate(cfg_keys):
    ax1.plot(results[k]["history"], color=cols[i], lw=1.4,
             label=list(COLORS.keys())[i], alpha=0.9)
ax1.set_title("Training Loss (MSE on PCA scores)", fontsize=11, color="#e2e8f0", pad=8)
ax1.set_xlabel("Epoch"); ax1.set_ylabel("MSE")
ax1.legend(framealpha=0.15, facecolor="#1a1d2e", fontsize=8)
ax1.grid(True, alpha=0.3); ax1.set_facecolor("#0d1117")

# ── Row 1 right: PCA variance explained ──────────────────────────────────────
ax1b = fig.add_subplot(gs[1, 2:])
pca_full = PCA(n_components=min(20, X_train_raw.shape[1])).fit(X_train_raw)
cumvar   = np.cumsum(pca_full.explained_variance_ratio_)
ax1b.plot(range(1, len(cumvar) + 1), cumvar, color="#00e5ff", lw=2, marker="o",
          markersize=4)
ax1b.axvline(N_MODES, color="#f59e0b", lw=1.5, ls="--",
             label=f"{N_MODES} components used ({var_explained:.1%})")
ax1b.axhline(0.99, color="#22c55e", lw=1, ls=":", alpha=0.6, label="99% threshold")
ax1b.set_xlabel("Number of PCA components")
ax1b.set_ylabel("Cumulative variance explained")
ax1b.set_title("Vol Surface PCA -- Variance Explained", fontsize=11, color="#e2e8f0", pad=8)
ax1b.legend(framealpha=0.15, facecolor="#1a1d2e", fontsize=8)
ax1b.grid(True, alpha=0.3); ax1b.set_facecolor("#0d1117")
ax1b.set_ylim(0, 1.05)

# ── Row 2: predicted vs actual vol surface (best config, first test day) ──────
best_cfg  = min(results, key=lambda k: results[k]["metrics_te"]["rmse"])
best_pred = results[best_cfg]["pred_surface"]
true_surf = true_surface_test

# Parse tenors and maturities from column names
import re
tenors     = sorted(set(int(re.search(r"Tenor : (\d+)", c).group(1))   for c in VOL_COLS))
maturities = sorted(set(float(re.search(r"Maturity : ([\d.]+)", c).group(1)) for c in VOL_COLS))
T, M       = len(tenors), len(maturities)

# Reshape day 0 of test into (T, M) grid
def to_grid(surface_row):
    grid = np.zeros((T, M))
    for idx, col in enumerate(VOL_COLS):
        t = int(re.search(r"Tenor : (\d+)", col).group(1))
        m = float(re.search(r"Maturity : ([\d.]+)", col).group(1))
        ti = tenors.index(t)
        mi = maturities.index(m)
        grid[ti, mi] = surface_row[idx]
    return grid

DAY = 0
pred_grid  = to_grid(best_pred[DAY])
true_grid  = to_grid(true_surf[DAY])
error_grid = pred_grid - true_grid

vmin = min(pred_grid.min(), true_grid.min())
vmax = max(pred_grid.max(), true_grid.max())

ax2a = fig.add_subplot(gs[2, :2])
im = ax2a.imshow(pred_grid, aspect="auto", cmap="plasma", vmin=vmin, vmax=vmax)
ax2a.set_title(f"Predicted Vol Surface ({best_cfg}) -- Test Day {DAY+1}",
               fontsize=10, color="#e2e8f0", pad=6)
ax2a.set_xlabel("Maturity"); ax2a.set_ylabel("Tenor")
ax2a.set_xticks(range(M)[::2]); ax2a.set_xticklabels([f"{m}" for m in maturities[::2]], fontsize=6, rotation=45)
ax2a.set_yticks(range(T)); ax2a.set_yticklabels([str(t) for t in tenors], fontsize=7)
plt.colorbar(im, ax=ax2a, fraction=0.03).set_label("Implied Vol", color="#94a3b8")

ax2b = fig.add_subplot(gs[2, 2:])
im2 = ax2b.imshow(true_grid, aspect="auto", cmap="plasma", vmin=vmin, vmax=vmax)
ax2b.set_title(f"True Vol Surface -- Test Day {DAY+1}",
               fontsize=10, color="#e2e8f0", pad=6)
ax2b.set_xlabel("Maturity"); ax2b.set_ylabel("Tenor")
ax2b.set_xticks(range(M)[::2]); ax2b.set_xticklabels([f"{m}" for m in maturities[::2]], fontsize=6, rotation=45)
ax2b.set_yticks(range(T)); ax2b.set_yticklabels([str(t) for t in tenors], fontsize=7)
plt.colorbar(im2, ax=ax2b, fraction=0.03).set_label("Implied Vol", color="#94a3b8")

# ── Row 3: error map + time series for one cell ───────────────────────────────
ax3a = fig.add_subplot(gs[3, :2])
emax = np.abs(error_grid).max()
im3  = ax3a.imshow(error_grid, aspect="auto", cmap="RdBu_r",
                   vmin=-emax, vmax=emax)
ax3a.set_title(f"Prediction Error (Pred - True) -- Test Day {DAY+1}",
               fontsize=10, color="#e2e8f0", pad=6)
ax3a.set_xlabel("Maturity"); ax3a.set_ylabel("Tenor")
ax3a.set_xticks(range(M)[::2]); ax3a.set_xticklabels([f"{m}" for m in maturities[::2]], fontsize=6, rotation=45)
ax3a.set_yticks(range(T)); ax3a.set_yticklabels([str(t) for t in tenors], fontsize=7)
plt.colorbar(im3, ax=ax3a, fraction=0.03).set_label("Error", color="#94a3b8")

# Time series for a single (Tenor=5, Maturity=5) cell over all test days
ax3b = fig.add_subplot(gs[3, 2:])
CELL_COL = "Tenor : 5; Maturity : 5"
cell_idx  = VOL_COLS.index(CELL_COL)
n_days    = min(100, len(true_surf))
ax3b.plot(true_surf[:n_days, cell_idx], color="#00e5ff", lw=1.5,
          label="True", alpha=0.9)
ax3b.plot(best_pred[:n_days, cell_idx], color="#f59e0b",  lw=1.5,
          label=f"QRC ({best_cfg})", alpha=0.9, ls="--")
ridge_cell = ridge_pred_surface[:n_days, cell_idx]
ax3b.plot(ridge_cell, color="#ef4444", lw=1.2,
          label="Ridge baseline", alpha=0.7, ls=":")
ax3b.set_title(f"Time Series: {CELL_COL} -- first {n_days} test days",
               fontsize=10, color="#e2e8f0", pad=6)
ax3b.set_xlabel("Test day index"); ax3b.set_ylabel("Implied Vol")
ax3b.legend(framealpha=0.15, facecolor="#1a1d2e", fontsize=8)
ax3b.grid(True, alpha=0.3); ax3b.set_facecolor("#0d1117")

plt.savefig("qrc_swaptions_dashboard.png", dpi=150, bbox_inches="tight",
            facecolor=fig.get_facecolor())
plt.show()
print("\nDashboard saved as qrc_swaptions_dashboard.png")

# =============================================================================
# 8) Final Summary
# =============================================================================
print("\n" + "="*65)
print("  SWAPTION VOL SURFACE FORECASTING SUMMARY")
print("="*65)
print(f"  {'Config':<22} {'Train RMSE':>11}  {'Test RMSE':>10}  {'Test R2':>8}")
print("-"*65)
for k in cfg_keys:
    m_tr = results[k]["metrics_tr"]
    m_te = results[k]["metrics_te"]
    flag = " *" if k == best_cfg else ""
    print(f"  {k:<22} {m_tr['rmse']:>11.6f}  {m_te['rmse']:>10.6f}  "
          f"{m_te['r2']:>8.4f}{flag}")
print("-"*65)
print(f"  {'Ridge baseline':<22} {'--':>11}  "
      f"{ridge_metrics['rmse']:>10.6f}  {ridge_metrics['r2']:>8.4f}")
print("="*65)
print(f"\n  Best config : {best_cfg}")
best_m = results[best_cfg]["metrics_te"]
if best_m["rmse"] < ridge_metrics["rmse"]:
    improvement = (ridge_metrics["rmse"] - best_m["rmse"]) / ridge_metrics["rmse"] * 100
    print(f"  QRC outperforms Ridge by {improvement:.1f}% RMSE reduction")
else:
    gap = best_m["rmse"] - ridge_metrics["rmse"]
    print(f"  QRC within {gap:.6f} RMSE of Ridge baseline")
    print(f"  Tip: increase N_MODES (try 6 or 8) for more PCA components")
print("="*65)
```
